```python
import re

# Función para tokenizar
def tokenize(text):
    text = text.lower()  # Convertir a minúsculas
    text = re.sub(r'[^\w\s]', '', text)  # Eliminar puntuación
    tokens = text.split()
    filtred_tokens = [word for word in tokens if len(word) >= 5]
    return filtred_tokens
```

**cramos nuestro diccionario de docimentos clave -> valor, titulo -> resumenes**

```python
documents = {}

for i in range(len(titulos)):
    documents[titulos[i]] = resumenes[i]
# 
```

**Diccionario con las palabras por documento**

```python
word_dict_per_doc = {}

for key, text in documents.items():
    word_set = set(tokenize(text))  # Tokenizar cada documento por separado
    word_list = sorted(word_set)  # ordenar
    
    word_dict_per_doc[key] = {idx + 1: word for idx, word in enumerate(word_list)}  # Crear diccionario
```

```python
# Crear un diccionario global con todas las palabras únicas de los tres documentos
global_word_set = set()

# Agregar todas las palabras únicas de cada documento al conjunto global
for text in documents.values():
    global_word_set.update(tokenize(text))

# Ordenar y asignar un índice a cada palabra en el diccionario global
global_word_list = sorted(global_word_set)
global_word_dict = {idx: word for idx, word in enumerate(global_word_list)}

# Imprimir el diccionario global de palabras únicas
print("\nDiccionario global de palabras únicas:")
for index, word in global_word_dict.items():
    print(f"{index}: {word}")

    
```

**paso 2 decomponer la matriz A en U, S y V**
**A = USV^T**

```python
# Descomposición en valores singulares (SVD) de la matriz A
U, S, VT = np.linalg.svd(A, full_matrices=False)
# Obtener la matriz V a partir de la transpuesta de V^T
V = VT.T
# Convertir S en una matriz diagonal
S_matrix = np.diag(S)

# Mostrar las matrices resultantes
print("Matriz U:")
print(U.shape)
print("\nMatriz S:")
print(S.shape)
# Mostrar la matriz V
print("\nMatriz V:")
print(V.shape)
print("\nMatriz V^T:")
print(VT.shape)
print("\nMatriz V^T:")
print(S_matrix.shape)
```

```python
umbral = 0.80

# Energía total (suma de los cuadrados de los valores singulares)
energia_total = np.sum(S**2)
    
# Energía acumulada
energia_acumulada = np.cumsum(S**2)
   
# Encontrar el menor k tal que la energía acumulada supere el umbral
k = np.searchsorted(energia_acumulada / energia_total, umbral) + 1

print(k)
```

```python

# Definir la consulta
query_text = "what is ontology?"
query_tokens = tokenize(query_text)
q = np.zeros((len(global_word_list), 1), dtype=int)
query_counts = Counter(query_tokens)
for i, word in enumerate(global_word_list):
    q[i, 0] = query_counts.get(word, 0)  # Asignar la frecuencia de la palabra
# Paso 5: Calcular las nuevas coordenadas de la consulta en el espacio reducido
# Se usa la ecuación q' = q^T U_k S_k^(-1)

# Invertir la matriz S_k
S_k_inv = np.linalg.inv(S_k)

# Transformar el vector de consulta
q_reduced = (q.T @ U_k) @ S_k_inv

# Mostrar los resultados
print("Coordenadas de los documentos en el espacio reducido (Paso 4):")
print(doc_coordinates)

print("\nNueva representación de la consulta en el espacio reducido (Paso 5):")
print(q_reduced)
```

```python
# Paso 6: Ordenar los documentos según la similitud del coseno en orden descendente

from sklearn.metrics.pairwise import cosine_similarity

# Calcular la similitud del coseno entre la consulta reducida y los documentos reducidos
similarity_scores = cosine_similarity(q_reduced, doc_coordinates)  # q' vs documentos

print(similarity_scores)


# Crear un diccionario con los documentos y sus respectivas similitudes
doc_similarity = {doc: similarity_scores[0, i] for i, doc in enumerate(documents.keys())}

# Ordenar los documentos por similitud de coseno en orden descendente
ranked_docs = sorted(doc_similarity.items(), key=lambda x: x[1], reverse=True)

resultados = []
# Mostrar los resultados
print("Ranking de documentos según similitud de coseno:")
for rank, (doc, similarity) in enumerate(ranked_docs, start=1):
    print(f"{rank}. {doc} - Similitud: {similarity:.4f}")
    resultados.append(f"{rank}. {doc} - Similitud: {similarity:.4f}")

# Mostrar el ranking final en el formato esperado
ranking_str = " > ".join([doc for doc, _ in ranked_docs])
print(f"\nOrden final de documentos: {ranking_str}")
```

```python
import langid
from googletrans import Translator

def detectar_idioma(texto):
    idioma, _ = langid.classify(texto)  # Detecta el idioma
    if idioma == "es":
        return "Español"
    elif idioma == "en":
        return "Inglés"
    else:
        return f"Otro idioma detectado: {idioma}"
    
def traducir_texto(texto, destino="en"):
    """
    Traduce un texto al idioma destino (por defecto, inglés).
    
    Args:
        texto (str): Texto a traducir.
        destino (str): Código del idioma destino (ejemplo: 'es' para español, 'en' para inglés).
    
    Returns:
        str: Texto traducido.
    """
    translator = Translator()
    traduccion = translator.translate(texto, dest=destino)
    return traduccion.text
```

```python
import tkinter as tk 
from tkinter import ttk, messagebox

class BuscadorApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Buscador de Documentos")
        self.root.geometry("600x400")
        
        # Marco para la búsqueda
        frame_busqueda = ttk.Frame(self.root, padding="10 10 10 10")
        frame_busqueda.pack(fill=tk.X)
        
        # Etiqueta y entrada de texto
        ttk.Label(frame_busqueda, text="Buscar:").pack(side=tk.LEFT, padx=(0, 5))
        self.entrada_busqueda = ttk.Entry(frame_busqueda, width=50)
        self.entrada_busqueda.pack(side=tk.LEFT, fill=tk.X, expand=True)
        
        # Botón de búsqueda
        self.boton_buscar = ttk.Button(frame_busqueda, text="Buscar", command=self.buscar)
        self.boton_buscar.pack(side=tk.LEFT, padx=(5, 0))
        
        # Lista para mostrar resultados
        frame_resultados = ttk.Frame(self.root, padding="10 10 10 10")
        frame_resultados.pack(fill=tk.BOTH, expand=True)
        
        self.lista_resultados = tk.Listbox(frame_resultados, height=15)
        self.lista_resultados.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
        
        # Barra de desplazamiento
        scrollbar = ttk.Scrollbar(frame_resultados, orient=tk.VERTICAL, command=self.lista_resultados.yview)
        scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
        self.lista_resultados.config(yscrollcommand=scrollbar.set)

        # Asociar evento de doble clic para abrir el archivo
        self.lista_resultados.bind("<Double-Button-1>", self.abrir_archivo)

        
        
    def buscar(self):
        print("boton buscar")

        consulta = self.entrada_busqueda.get().strip()

        print(consulta)

        if not consulta:
            messagebox.showwarning("Entrada Vacía", "Por favor, ingresa un término de búsqueda.")
            return
        

        if detectar_idioma(consulta) == "Español":
            consulta = traducir_texto(consulta, "en")
        
        print(consulta)


        resultados = query_for_UI(consulta)

        
        # Limpiar la lista actual
        self.lista_resultados.delete(0, tk.END)
        
        if resultados:
           for i in range(len(resultados)):
               self.lista_resultados.insert(tk.END,resultados[i])
        else:
            self.lista_resultados.insert(tk.END, "No se encontraron resultados.")
        
    def abrir_archivo(self, event=None):
        seleccion = self.lista_resultados.curselection()  # Obtiene la selección actual
        print(seleccion)
        if seleccion:
            pelicula = self.lista_resultados.get(seleccion)  # Obtiene el nombre de la película seleccionada
            print(pelicula)
            resumen = diccionario_peliculas.get(pelicula, "Resumen no disponible.")  # Obtiene el resumen
            print(resumen)
            messagebox.showinfo(pelicula, resumen)  #
            

  

```

```python

def main():
    root = tk.Tk()
    app = BuscadorApp(root)
    root.mainloop()


if __name__ == "__main__":
    main()
```
